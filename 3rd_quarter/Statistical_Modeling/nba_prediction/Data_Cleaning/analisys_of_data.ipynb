```python
new_nba_df = nba_df
not_usa = new_nba_df.iloc[:, 5] == 'USA'

new_nba_df = new_nba_df.loc[not_usa]
new_nba_df
#oli jeje
```

```python
new_nba_df = new_nba_df.drop(new_nba_df.columns[0:7], axis = 1)
new_nba_df
```

```python
#Transforming season column values into numeric.
dict_season = {}
set_season = list(set(new_nba_df['season']))
for i in range(len(set_season)):
    dict_season[set_season[i]] = i+1

new_nba_df['numerated_season'] = new_nba_df['season'].apply(lambda x: dict_season[x])
new_nba_df['numerated_season']
new_nba_df[['season', 'numerated_season']]
```

```python
#Ordering seasons
assigned_seasons = set()
def get_season_number(season):
    if season not in assigned_seasons:
        assigned_seasons.add(season)
    return len(assigned_seasons)

new_nba_df['numerated_season'] = new_nba_df['season'].apply(get_season_number)
print(new_nba_df[['season', 'numerated_season']])
```

```python
#Setting the dataframe columns that we are going to use.
new_nba_df = new_nba_df.drop(new_nba_df.columns[-3:-2], axis = 1)
new_nba_df = new_nba_df.drop(new_nba_df.columns[3], axis = 1)
new_nba_df
#The dependent variable is net_orating and the independent ones are the other ones except numerated season.
```
